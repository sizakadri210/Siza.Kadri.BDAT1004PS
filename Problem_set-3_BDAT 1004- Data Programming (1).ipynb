## Question-4 (Wind Statistics)

#### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

#### Step 2. Import the dataset from the attached file wind.txt

```python
column_names = ['Yr', 'Mo', 'Dy', 'RPT', 'VAL', 'ROS', 'KIL', 'SHA', 'BIR', 'DUB', 'CLA', 'MUL', 'CLO', 'BEL', 'MAL']
pd.read_csv('wind.txt', sep='\s+', names=column_names, skiprows=1)
```

#### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index

```python
column_names = ['Yr', 'Mo', 'Dy', 'RPT', 'VAL', 'ROS', 'KIL', 'SHA', 'BIR', 'DUB', 'CLA', 'MUL', 'CLO', 'BEL', 'MAL']
data=pd.read_csv('wind.txt', sep='\s+', names=column_names, skiprows=1)
data['DateTime'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1), format='%y-%m-%d')
data.set_index('DateTime', inplace=True)
data.drop(['Yr', 'Mo', 'Dy'], axis=1, inplace=True)
print(data)
```

#### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
wind = pd.read_csv('wind.txt', sep='\s+', names=column_names, skiprows=1)
def fix_year(year):
    if year >= 100:
        return year
    elif year >= 23:
        return 1900 + year
    else:
        return 2000 + year
wind['Yr'] = wind['Yr'].apply(fix_year)
print(wind)
```

#### Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
wind['Date'] = pd.to_datetime(wind[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1), format='%Y-%m-%d')
wind.set_index('Date', inplace=True)
wind.drop(['Yr', 'Mo', 'Dy'], axis=1, inplace=True)
print(wind)
print("Datatype:",data.index.dtype)
```

#### Step 6. Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below.

```python
missing_value=wind.isnull().sum()
print("Missing value for each location over entire record:",missing_value)
total_missing_value= wind.isnull().sum().sum()
print("Total missing values in entire record:",total_missing_value)
```

#### Step 7. Compute how many non-missing values there are in total.

```python
total_non_missing_value=wind.notnull().sum().sum()
print("Non-missing values in total:",total_non_missing_value)
```

## Question-6(US marriages-divorces-1867-2014) 
Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv('us-marriages-divorces-1867-2014.csv')
marriages_per_capita =data['Marriages_per_1000'] 
divorces_per_capita =data['Divorces_per_1000'] 
plt.plot(data['Year'], marriages_per_capita, label='Marriages per Capita', color='red')
plt.plot(data['Year'], divorces_per_capita, label='Divorces per Capita', color='green')
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.legend()
plt.show()
```

## Question-8(Actor_kill_counts)
Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!

```python
data=pd.read_csv('actor_kill_counts.csv')
sorted_actors = data.sort_values(by='Count', ascending=False)
plt.barh(sorted_actors['Actor'], sorted_actors['Count'])
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood')
for i, (actor, kill_count) in enumerate(zip(sorted_actors['Actor'], sorted_actors['Count'])):
    plt.text(kill_count + 1, i, f'{kill_count}', va='center', color='black')
plt.show()
```

## Question-9(Roman-emperor-regins)
Create a pie chart showing the fraction of all Roman Emperors that were assassinated.


```python
data=pd.read_csv("roman-emperor-reigns.csv")
data.head()
```

```python
import matplotlib.pyplot as plt
total_emperor=len(data)
assassinated_count = data[data['Cause_of_Death'].str.contains('Assassinated')].shape[0]
non_assassinated_count = total_emperor - assassinated_count
labels = ['Assassinated', 'Non-Assassinated']
sizes = [assassinated_count, non_assassinated_count]
colors = ['red', 'green']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  
plt.title("Fraction of Roman Emperors Assassinated")
plt.show()
```

## Question-10(Arcade-revenue-vs-cs-doctorates)
Create a scatter plot showing the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 
2000 and 2009.
Don't forget to label your axes!
Color each dot according to its year

```python
data=pd.read_csv('arcade-revenue-vs-cs-doctorates (1).csv')
for year in range(2000, 2010):
    year_data = data[data['Year'] == year]
    plt.scatter(year_data['Total Arcade Revenue (billions)'], year_data['Computer Science Doctorates Awarded (US)'],label=str(year))
plt.xlabel('Total Revenue')
plt.ylabel('Number of Computer Science PhDs')
plt.title('Relationship between Arcade Revenue and Computer Science PhDs (2000-2009)')
plt.legend()
plt.show()
```
